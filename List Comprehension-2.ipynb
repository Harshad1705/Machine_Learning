```python
# Get the index and value as a tuple for items in the list

items = [ 'Hi' , 1.29 , 3, 4,(0.1,'hi')]

l=[(index,value) for index,value in enumerate(items)]
print(l)
```

```python
for i in range(0,len(items)-1):
  t = (i,items[0])
  print(t)
```

```python
# Find the common number using list comprehension

l1=[1,2,3,4,5]
l2=[1,2,3,4,5,6,7,8,9]

l=[ i for i in l1 if(i in l2)]

print(l)

l0 = [ i for i in l1 for j in l2 if i==j]
print(l0)
```

```python
# Count the number of letters

sentence = "I am Harshad"

l = [ i for i in sentence if i!=' ']
print(len(l))

space = sentence.split()
print(space)

l0=[ i for i in sentence.split()]
print(len(l0))
```

```python
# Count the number of word which are not alphanumberic

s =" This is CSIT 1984"

l=[ i for i in s.split() if i.isalpha() ]
print(len(l))


# Count the numbers of words character greater than4
l=[ i for i in s.split() if i.isalpha() and len(i)>=4]
print(len(l))
```

```python
# Use a nested list comprehension to find all the numbers in range 1 to 100 that are divisivblw by 2 to 9


l = [i for i in range(1,101) for j in range(2,10)  if i%j==0  ]
print(l)


l2=[]

for i in l:  
  if i not in l2:
    l2.append(i)

print(l2)
  





```
